# Seaborn圖樣式
有五種預設seaborn主題：darkgrid，whitegrid，dark，white，和ticks。它們各自適合不同的應用和個人喜好。默認主題是darkgrid

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#設定圖形樣式 - whitegrid
# use sns.set
sns.set(style = 'whitegrid')
# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
'''
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
'''
rs = np.random.RandomState()
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
# 畫圖
# sns.residplot
sns.residplot(x, y, color = 'b')
```

## 繪製單變數分佈: 
快速檢視多個單變數分佈的最便捷方式是distplot() 函數。默認情況下,這將繪製直方圖並適合內核密度估計值(KDE)。

作業目標(2)
(1)使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位,但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵:
(2)有無kde對圖形分布的影響

```python
# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Question 2-(1)')

sns.set(color_codes=False)

x = np.random.normal(size=100)

sns.distplot(x, ax=axes[0])
axes[0].set_title('estimates bin numbers by built-in function')

sns.distplot(x, bins=3, ax=axes[1])
axes[1].set_title('with lesser bin numbers')

sns.distplot(x, bins=50, ax=axes[2])
axes[2].set_title('with more bin numbers')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
fig.suptitle('Question 2-(2)')

sns.set(color_codes=False)

x = np.random.normal(size=100)

sns.distplot(x, ax=axes[0])
axes[0].set_title('estimates bin numbers by built-in function')

sns.distplot(x, ax=axes[1], kde=False)
axes[1].set_title('without kde')

sns.distplot(x, bins=50, ax=axes[2])
axes[2].set_title('with more bin numbers')

sns.distplot(x, bins=50, ax=axes[3], kde=False)
axes[3].set_title('without kde')
```

```python
print('with more bins, the visualization of data is not like normal distribution')
print('without kde fitting, the axis is more compact and close to the data point distribution')
```
